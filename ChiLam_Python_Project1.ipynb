```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("https://raw.githubusercontent.com/CunyLaguardiaDataAnalytics/datasets/master/2014-15_To_2016-17_School-_Level_NYC_Regents_Report_For_All_Variables.csv")

# Step 1: Clean the data
# Replace non-numeric values in the 'Mean Score' column with NaN
df['Mean Score'] = pd.to_numeric(df['Mean Score'], errors='coerce')

# Drop rows with missing values in the 'Mean Score' and 'Regents Exam' columns
data_cleaned = df.dropna(subset=['Mean Score', 'Regents Exam'])

# Step 2: Filter high schools in Manhattan
manhattan_high_schools = data_cleaned[(data_cleaned['School Level'] == 'High school') & (data_cleaned['School DBN'].str[2] == 'M')]

# Step 3: Choose a specific school in Manhattan to compare
specific_school_dbn = '02M475'  #Stuyvesant High School

# Filter data for the specific school
specific_school_data = manhattan_high_schools[manhattan_high_schools['School DBN'] == specific_school_dbn]

# Extract the mean score for the specific school
specific_school_mean_score = specific_school_data['Mean Score'].iloc[0]

# Calculate the mean score for all high schools in Manhattan
mean_score_all_manhattan = manhattan_high_schools['Mean Score'].mean()

print("\n\033[1mAnalysis of Stuyvesant High School's Performance in Regents Exams Compared to Manhattan High Schools\033[0m".center(70))

# Step 4: Descriptive Statistics
print("\n\033[1mDescriptive Statistics:\033[0m\n")
print("Specific School (Stuyvesant H.S.):")
print(specific_school_data['Mean Score'].describe())
print("\nAll H.S. in Manhattan:")
print(manhattan_high_schools['Mean Score'].describe())
print ()

# Step 5: Create a bar plot to compare mean scores
plt.figure(figsize=(4.5, 5.5))
plt.ylim(0, 100)  # Set y-axis limit to ensure consistency
plt.bar(['Stuyvesant H.S.', 'All H.S. in Manhattan'], [specific_school_mean_score, mean_score_all_manhattan], color=['skyblue', 'salmon'])

# Add labels to the bars
plt.text(0, specific_school_mean_score, round(specific_school_mean_score, 2), ha='center', va='bottom')
plt.text(1, mean_score_all_manhattan, round(mean_score_all_manhattan, 2), ha='center', va='bottom')

plt.title("Stuyvesant H.S. vs All H.S in Manhattan")
plt.xlabel('School Group')
plt.ylabel('Mean Score: Regents Exam')
plt.show()

# Step 6: Conclusion
print("\n\033[1mConclusion Statement:\033[0m")
if specific_school_mean_score > mean_score_all_manhattan:
    print("\nStuyvesant H.S. has a higher Regents Exam mean score compared to all high schools in Manhattan.")
    print("This suggests that Stuyvesant H.S. performs well academically compared to all other high schools in the borough.")
elif specific_school_mean_score < mean_score_all_manhattan:
    print("\nStuyvesant H.S. has a lower Regents Exam mean score compared to all high schools in Manhattan.")
    print("This indicates that Stuyvesant may have areas for improvement in academic performance.")
else:
    print("\nStuyvesant H.S. has the same Regents Exam mean score as all high schools in Manhattan.")
    print("This implies that Stuyvesant's academic performance is consistent with the average performance of high schools in the borough.")

```
